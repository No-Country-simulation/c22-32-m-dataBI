## Perfiles de clientes


```python
# Filtrar solo los eventos de compra
compras = ecommerce[ecommerce['event_type'] == 'purchase']

# Contar compras por usuario
compras_por_usuario = compras.groupby('user_id')['event_type'].count().reset_index()

# Renombrar la columna de conteo
compras_por_usuario.rename(columns={'event_type': 'purchase_count'}, inplace=True)
```

```python
# Clasificar a los usuarios según el número de compras
def tipo_usuario(purchase_count):
    if purchase_count >= 10:
        return 'Frecuente'
    elif 3 <= purchase_count < 10:
        return 'Moderado'
    elif 1 <= purchase_count < 3:
        return 'Ocasional'
    else:
        return 'Inactivo'

# Aplicar la función para categorizar usuarios
compras_por_usuario['user_category'] = compras_por_usuario['purchase_count'].apply(tipo_usuario)
```

```python
# Unir las categorías al DataFrame principal
ecommerce = ecommerce.merge(compras_por_usuario, on='user_id', how='left')

# Completar valores nulos con 'Inactivo' para usuarios sin compras
ecommerce['user_category'].fillna('Inactivo', inplace=True)
```

```python
# Contar usuarios por categoría
category_counts = ecommerce['user_category'].value_counts()

# Gráfico de barras
sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")
plt.title('Distribución de Usuarios por Frecuencia')
plt.xlabel('Tipo de Usuario')
plt.ylabel('Cantidad de Usuarios')
plt.show()
```

```python
# Calcular el gasto promedio por categoría
avg_spend_by_category = ecommerce.groupby('user_category')['price'].mean()


# Gráfico de barras para el gasto promedio
avg_spend_by_category.plot(kind='bar', color='skyblue')
plt.title('Gasto Promedio por Tipo de Usuario')
plt.xlabel('Tipo de Usuario')
plt.ylabel('Gasto Promedio')
plt.show()
```

```python
categorias = ecommerce[ecommerce['event_type'] == 'purchase']['category_code'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))
categorias.plot(kind='bar', color='purple')
plt.title("Análisis de Categorías (Top 10)", fontsize=16)
plt.xlabel("Categoría", fontsize=12)
plt.ylabel("Cantidad de Compras", fontsize=12)
plt.show()
```

## Ventas diarias

```python
# Agrupación de datos por día y producto
ventas_diarias = compras.groupby(['date', 'product_id']).agg({
    'price': 'sum',  # Ventas totales por producto
    'user_id': 'nunique',  # Número de usuarios únicos que compraron
}).reset_index()

# Renombrar columnas
ventas_diarias.rename(columns={'price': 'total_sales', 'user_id': 'unique_buyers'}, inplace=True)
```

```python
# 1. Agregar medias móviles (ventas promedio en ventanas temporales)
ventas_diarias['7_day_avg_sales'] = ventas_diarias.groupby('product_id')['total_sales'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# 2. Agregar tendencia (diferencia entre las ventas de hoy y las de ayer)
ventas_diarias['sales_trend'] = ventas_diarias.groupby('product_id')['total_sales'].diff()

# 3. Agregar lag features (ventas de días anteriores como predictores)
for lag in range(1, 8):
    ventas_diarias[f'lag_{lag}_sales'] = ventas_diarias.groupby('product_id')['total_sales'].shift(lag)
```

```python
ventas_dia = ecommerce[ecommerce['event_type'] == 'purchase'].groupby('date')['price'].sum()

plt.figure(figsize=(12, 6))
plt.plot(ventas_dia.index, ventas_dia.values, marker='o', color='blue')
plt.title("Ventas Totales por Día", fontsize=16)
plt.xlabel("Fecha", fontsize=12)
plt.ylabel("Ingresos Totales", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

## PKI Optimización de la Tasa de Conversión

Incrementar la tasa de conversión en un 7% pasando de 11.34% a 12.13%.
Metrica principal: Tasa de Conversión
Acciones de mejora:
Identificación de puntos de fricción en cada etapa: Análisis del embudo para conocer puntos de abandono.
posibles encuestas para conocer las razones de abandono
Simplificación de procesos: Menos pasos entre el inicio y la conversión.-clics + ventas.
Fidelización: Convertir a los clientes en recurrentes.
agradecimientos,descuentos exclusivos.
Urgencia y escasez: Aplica elementos psicológicos (como ofertas limitadas) para incentivar la acción inmediata.

```python
# Tasa de abandono de carrito
usuarios_carrito = ecommerce[ecommerce['event_type'] == 'cart']['user_id'].nunique()
tasa_abandono_carrito = (usuarios_carrito - usuarios_que_compran) / usuarios_carrito
print("Tasa de Abandono de Carrito:", tasa_abandono_carrito)
```

```python
conversion = ecommerce['event_type'].value_counts()
steps = ['view', 'cart', 'purchase']
conversion_funnel = [conversion.get(step, 0) for step in steps]

plt.figure(figsize=(8, 6))
plt.barh(steps, conversion_funnel, color=['blue', 'orange', 'green'])
plt.title("Embudo de Conversión", fontsize=16)
plt.xlabel("Número de Eventos", fontsize=12)
plt.ylabel("Etapas", fontsize=12)
plt.show()
```

## Tasa de abandono segun categoria

```python
abandono_categoria = ecommerce[ecommerce['event_type'] == 'cart'].groupby('category_code')['user_id'].nunique()
compra_categoria = ecommerce[ecommerce['event_type'] == 'purchase'].groupby('category_code')['user_id'].nunique()
tasa_abandono = (abandono_categoria - compra_categoria) / abandono_categoria

plt.figure(figsize=(10, 6))
tasa_abandono.nlargest(10).plot(kind='bar', color='red')
plt.title("Tasa de Abandono por Categoría (Top 10)", fontsize=16)
plt.xlabel("Categoría", fontsize=12)
plt.ylabel("Tasa de Abandono", fontsize=12)
plt.show()
```
